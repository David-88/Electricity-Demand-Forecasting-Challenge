## 2.1 首先复习matplotlib+numpy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 先复习matplotlib

#Normally 需要 %matplotlib inline --jupyter notebook不需要

#创建(-5,5)一共101个均匀分布数据
x1=np.linspace(-5,5,101)
# y=sin(x)
y1=np.sin(x1)
# 绘制图像
plt.plot(x1,y1)
```

```python
# 实际上我们先创建画布fig，再引入图形ax
fig,ax=plt.subplots()
# title
ax.set_title("sin")
# x轴标识
ax.set_xlabel("rad")
# 绘制
ax.plot(x1,y1)
```

```python
# 绘制各种图形
# 取[0,100)->0-99的整数数组
x2=np.arange(100)
# random.rand(100) 表示0-1范围内选取100个随机数数据
y2=x2 * np.random.rand(100)

# 绘制散点图
plt.scatter(x2,y2)
```

```python
# 绘制直方图
# 参数 bins 指定了直方图中条形的数量
# 图中y表示落在x指定区间范围内的数字的数量-频数
plt.hist(y2,bins=5)
```

```python
# 柱状图
plt.bar(x2,y2)
```

```python
# 折线图
plt.plot(x2,y2)
```

```python
# 箱形图
plt.boxplot(y2)
```

### 下面利用红酒数据集

```python
# 从sklearn中获取
from sklearn.datasets import load_wine
# 加载红酒数据集
data=load_wine()
```

```python
# 取出index=0的alcoho所有数据
x3=data.data[:,[0]]
# 取出index=9的color_intensity所有数据
y3=data.data[:,[9]]

# 绘制散点图
plt.scatter(x3,y3)
```

```python
# 绘制直方图
plt.hist(y3,bins=5)
```

### 2.2使用pandas理解和处理数据

```python
# 将load_wine数据集转化为pandas的dataframe格式
# 内容 data.data, 列名 依旧~
df_X=pd.DataFrame(data.data,columns=data.feature_names)
# 显示前5行数据
df_X.head()
```

```python
# 将load_wine数据集中目标变量转化为dataframe
df_Y=pd.DataFrame(data.target,columns=["kind(target)"])
# 查看前5行
df_Y.head()
```

```python
# 合并dataframe
# axis=0: 按行（row）进行合并，即垂直方向。
# axis=1: 按列（column）进行合并，即水平方向。
df=pd.concat([df_X,df_Y],axis=1)
# 查看前5行
df.head()
```

### 接下来通过可视化和数理统计来分析这里的数据

```python
# 绘制直方图,选择所有行，并且只选取名为 "alcohol" 的这一列。
# default bins=10
plt.hist(df.loc[:,"alcohol"])
```

```python
# 绘制箱形图
plt.boxplot(df.loc[:,"alcohol"])
```

```python
# 输出相关系数
df.corr()
```

```python
# 输出每一列的统计信息
df.describe()
```

```python
# 将所有列之间的关系进行可视化
from pandas.plotting import scatter_matrix
a=scatter_matrix(df,figsize=(15,15))
```
